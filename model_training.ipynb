# Training the models for image classification

```python
from utils import fix_randomness, load_cifar10_trainset, load_cifar10_testset, generate_dataloader, train_model, test_model
from models.models import get_model_class
```

```python
fix_randomness(12345)
cifar10_trainset = load_cifar10_trainset()
cifar10_trainloader = generate_dataloader(cifar10_trainset)
```

```python
configs = {
    "device": "cuda:0",
    "lr": 0.001,
    "momemtum": 0.9,
    "num_epochs": 50,
    "save_dir": "cifar_net.pth",
}
model = get_model_class("Net")()
```

```python
model = train_model(model, cifar10_trainloader, configs)
```

```python
cifar10_testset = load_cifar10_testset()
cifar10_testloader = generate_dataloader(cifar10_testset)
```

```python
test_model(model, cifar10_testloader, configs)
```
